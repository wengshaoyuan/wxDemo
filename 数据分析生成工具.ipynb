```python
# encoding=utf-8
import pandas as pd
from pylab import *
import matplotlib
# from pylab import mpl
#设置字体为楷体
matplotlib.rcParams['font.sans-serif'] = ['KaiTi']

matplotlib.rc("font",family='YouYuan')
mpl.rcParams['font.sans-serif'] = ['SimHei']  # 添加这条可以让图形显示中文
mpl.rcParams['font.sans-serif'] = ['SimHei']
def areaMessage(path,sheet_name):
    data = pd.read_excel(path, sheet_name=sheet_name)

    data = data[['区县', '该工单是否扩容完成（林晓辉）', '归属类别（刘建峰）', '扩容方式（林素芳，魏奶生，林联必）']]
    dfg = data.groupby("区县")

    areaDict = {}
    for i in dfg:
        area = i[0]
        sectionData = i[1]

        if area not in areaDict.keys():
            areaDict[area] = {}

        KRcountNotDond = sectionData['该工单是否扩容完成（林晓辉）'].isna().sum()
        KRcountDond = sectionData['该工单是否扩容完成（林晓辉）'].count()
        areaDict[area]['扩容未完成'] = KRcountNotDond
        areaDict[area]['扩容已完成'] = KRcountDond

        classProperty = sectionData['归属类别（刘建峰）'].value_counts(dropna=False)
        for i in range(len(classProperty.index)):
            areaDict[area][classProperty.index[i]] = classProperty.values[i]

        KuoRongMethod = sectionData['扩容方式（林素芳，魏奶生，林联必）'].value_counts(dropna=False)
        for i in range(len(KuoRongMethod.index)):
            areaDict[area][KuoRongMethod.index[i]] = KuoRongMethod.values[i]

    df = pd.DataFrame(areaDict).T
    return df
data1=areaMessage("/home/wy/桌面/高负荷扩容小区进度跟踪表20210712.xlsx",'362个小区')
data2=areaMessage("/home/wy/桌面/高负荷扩容小区进度跟踪表20210713.xlsx",'362个小区')

data1=data1.fillna(0)
data2=data2.fillna(0)

print(data1)
areaLocation=list(data2.index)
compareClass=list(data2.columns)

def plts(names,y,y1,l):

    x = range(len(names))
   
    plt.plot(x, y, marker='o', mec='r', mfc='w', label=u'y=x^2曲线图')
    plt.plot(x, y1, marker='*', ms=10, label=u'y=x^3曲线图')
    plt.legend()  # 让图例生效
    plt.xticks(x, names, rotation=45)
    plt.margins(0)
    plt.subplots_adjust(bottom=0.15)
    plt.xlabel(u"time(s)邻居")  # X轴标签
    plt.ylabel("RMSE")  # Y轴标签
    plt.title(l)  # 标题

    plt.show()


for l in areaLocation:
    day1 = data1.loc[l]
    day2 = data2.loc[l]
    numberSequence1=[]
    numberSequence2 = []
    for c in compareClass:
        numberSequence1.append(day1.loc[c])
        numberSequence2.append(day2.loc[c])
   
    plts(compareClass,numberSequence1,numberSequence2,l)

```
